### Missing value Identification and Treatment

```python
tech_df.drop(['customer_id', 'order_id'], axis = 1, inplace = True)
tech_df.info()
```

#### **Question 3:** Are there any missing values in the data? If so, treat them using an appropriate method.

```python
# write you are code here 
tech_df.isna().sum()
```

```python
tech_df['education'].fillna(tech_df['education'].mode().values[0], inplace = True)
tech_df['delivery_charges'].fillna(tech_df['delivery_charges'].median(), inplace = True)
tech_df['order_type'].fillna(tech_df['order_type'].mode().values[0], inplace = True)
tech_df['order_payment'].fillna(tech_df['order_payment'].mode().values[0], inplace = True)
```

- The missing values under education, order_type and order_payment variables were treated by replacing the null values with the value which has the most frequency in each variable. This is because the variables are categorical in nature.
- The missing values under delivery_charges were treated by replacing the null values with the median of delivery_charges. We used median and not mean because it is not affected by extreme values which the delivery_charges variable has.

From the above, there are five variables (columns) - order_price, delivery_charges, coupon_discount, distance_to_nearest_warehouse and total_bill - that have outliers, and it needs to be treated.

#### Treating Order_price Outlier

```python
min_order_price, max_order_price = tech_df["order_price"].quantile([0.05, 0.95])
print(round(min_order_price, 2))
print(round(max_order_price, 2))
```

```python
tech_df["order_price"] = np.where(tech_df["order_price"] < min_order_price, min_order_price, tech_df["order_price"])
tech_df["order_price"] = np.where(tech_df["order_price"] > max_order_price, max_order_price, tech_df["order_price"])

tech_df[["order_price"]].boxplot(figsize = (5, 4))
```
